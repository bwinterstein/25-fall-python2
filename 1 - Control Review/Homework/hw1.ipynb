```python
def is_perfect(input_int):
    total = 0
    for numbers in range(1, input_int):
        if input_int % numbers == 0:
            total += numbers
    return total == input_int
```

```python
print(is_perfect(100))
print(is_perfect(496))
```

```python
def lcm(num1,num2):
    multiple = max(num1,num2)
    while multiple % num1 != 0 or multiple % num2 != 0:
        multiple += 1
    return multiple
```

```python
print(lcm(6,10))
print(lcm(3,10))
print(lcm(10,30))
```
