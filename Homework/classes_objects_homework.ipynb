```python
# Just as def defines a function but does not call it, 
# class defines an object but does not instantiate one
class Point:
    '''
    Represents a point in 2D space

    attributes: x, y
    '''

    # Constructor always requires self as the first argument, followed by whatever attributes you need
    # The attributes here have been assigned default values
    def __init__(self, input_x = 0, input_y = 0): #__init__, self, must be used everytime
        self.x = input_x
        self.y = input_y
```

```python
# One purpose of classes is to bundle together (encapsulate) related data so it can be
# reused in other classes/functions.
class Rectangle:
    '''
    Represents a rectangle with lower left corner at corner.

    attributes: corner (Point object), height (float), width (float)
    '''

    # PRACTICE: Define the constructor below
    def __init__(self,input_Point = Point(), input_height = 0, input_width = 0):
        self.corner = input_Point
        self.height = input_height
        self.width = input_width
```

```python
def get_corners(inputRect):
    input_x = inputRect.corner.x
    input_y = inputRect.corner.y
    input_height = inputRect.height
    input_width = inputRect.width 

    point1 = Point(input_x,input_y)
    point2 = Point(input_x + input_width, input_y)
    point3 = Point(input_x, input_y + input_height)
    point4 = Point(input_x + input_width, input_y + input_height)

    return point1, point2, point3, point4
```

```python
def distance(input_Point1, input_Point2):
    import math

    x_coord1 = input_Point1.x
    y_coord1 = input_Point1.y
    x_coord2 = input_Point2.x
    y_coord2 = input_Point2.y

    dist = math.sqrt((x_coord2 - x_coord1) ** 2 + (y_coord2 - y_coord1) ** 2)

    return dist
```
