````python
def fitness(knapsack, capacity):
    total_weight = sum(items[i][0] for i in knapsack)
    total_value = sum(items[i][1] for i in knapsack)
    if total_weight > capacity:
        return 0
    return total_value
````

````python
def mutate(parent_knapsack, capacity):
    new_knapsack = parent_knapsack.copy()

    available_items = [i for i in range(len(items)) if i not in new_knapsack]
    if available_items:
        item_to_add = random.sample(available_items, 1)[0]
        new_knapsack.append(item_to_add)

    while sum(items[i][0] for i in new_knapsack) > capacity and len(new_knapsack) > 0:
        item_to_remove = random.sample(new_knapsack, 1)[0]
        new_knapsack.remove(item_to_remove)

    return new_knapsack
````

````python
def knapsack(capacity):
    best_knapsack = []
    best_value = fitness(best_knapsack, capacity)
    print("Initial: Value", best_value, best_knapsack)

    same_count = 0
    generation = 0

    while same_count < 2: 
        generation += 1
        improvement_found = False

        for _ in range(100):
            offspring = mutate(best_knapsack, capacity)
            offspring_value = fitness(offspring, capacity)

            if offspring_value > best_value:
                best_knapsack = offspring
                best_value = offspring_value
                improvement_found = True

        print("Generation", generation, ": Value", best_value, best_knapsack)

        if improvement_found:
            same_count = 0
        else:
            same_count += 1

    print("\nTerminated after", generation, "generations.")
    return best_knapsack
````
