FUNCTION knapsack(capacity):
    Start with an empty bag
    Set best_value to 0
    Set same_count to 0

    WHILE same_count is less than 2:
        Make an empty list called offspring_list
        Repeat 100 times:
            Make a new bag by calling mutate(current_bag)
            Add that new bag to offspring_list
        Find the bag in offspring_list with the highest fitness
        IF that bag's fitness is better than best_value:
            Make that bag the new current bag
            Update best_value
            Reset same_count to 0
        ELSE:
            Add 1 to same_count
    RETURN the best bag found

FUNCTION mutate(bag):
    Make a copy of the bag
    Pick one random item that is not already in the bag
    Add that item to the bag
    WHILE the total weight of the bag is too heavy:
        Remove a random item from the bag
    RETURN the new bag

FUNCTION generate_next_generation(current_bag, items, capacity):
    Make an empty list offspring_list
    Repeat 100 times
        new_bag = mutate(current_bag)
        Add new_bag to offspring_list
    Find the bag in offspring_list with the highest FITNESS value
    RETURN that bag

FUNCTION fitness(bag):
    total_weight = add up weights of all items in the bag
    total_value = add up values of all items in the bag
    IF total_weight is more than the capacity:
        RETURN 0
    ELSE:
        RETURN total_value
