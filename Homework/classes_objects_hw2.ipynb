```python
class Point:
    '''
    Represents a point in 2D space

    attributes: x, y
    '''

    # Constructor always requires self as the first argument, followed by whatever attributes you need
    # The attributes here have been assigned default values
    def __init__(self, input_x = 0, input_y = 0): #__init__, self, must be used everytime
        self.x = input_x
        self.y = input_y

    def __str__(self):
        #s = '(' + self.x + ','  + self.y, ')' Slightly ugly syntax that does the same job
        return f'({self.x}, {self.y})'
    
class Circle:

    def __init__(self, center = Point(0,0), radius = 0):
        self.center = center
        self.radius = radius

    def __str__(self):
        return f"Center: {self.center}, Radius: {self.radius}"
    
    def get_area(self):
        import math

        return self.radius ** 2 * math.pi
    
    def to_square(self):
        import math

        circle_area = Circle.get_area(self)
        square_side = math.sqrt(circle_area)
        corner_x = self.center.x - square_side/2
        corner_y = self.center.y - square_side/2
        corner = Point(corner_x,corner_y)
        
        return Rectangle(corner,square_side,square_side)
    
class Rectangle:
    '''
    Represents a rectangle with lower left corner at corner.

    attributes: corner (Point object), height (float), width (float)
    '''

    # PRACTICE: Define the constructor below
    def __init__(self,input_Point, input_height, input_width):
        self.corner = input_Point
        self.height = input_height
        self.width = input_width

    def __str__(self):
        return f'Corner: {self.corner}, Width: {self.width}, Height: {self.height}'

    def get_area(self):
        return self.height * self.width
    
    def __lt__(self,other):
        if self.get_area() < other.get_area():
            return True
        return False
    
    def to_circle(self):
        import math

        area_rectangle = Rectangle.get_area(self)
        radius_circle = math.sqrt(area_rectangle/math.pi)

        middle_x = self.corner.x + (self.width/2)
        middle_y = self.corner.y + (self.height/2)
        circle_middle = Point(middle_x, middle_y)

        return Circle(circle_middle, radius_circle)
```

```python
rect = Rectangle(Point(),2,2)
print(rect.to_circle())
```
